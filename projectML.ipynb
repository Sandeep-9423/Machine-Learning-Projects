```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
info = pd.read_csv("Book.csv")
```

```python
info.head()
```

```python
infomain = info.drop(["Unnamed: 1","Unnamed: 5"],axis='columns')
```

```python
infomain.head()
```

```python
infomain.info()
```

```python
infomain.describe()
```

```python
infomain.isnull().sum()
```

```python
sns.barplot(x='Habits',y='charges',data=infomain)
```

```python
dumy = pd.get_dummies(infomain.Habits,dtype=int)
```

```python
dumy
```

```python
merging = pd.concat([infomain,dumy],axis='columns')
```

```python
merging
```

```python
end = merging.drop(['Habits'],axis="columns")
```

```python
end.head()
```

```python
end.corr()
```

```python
sns.heatmap(end.corr(),annot = True)
```

```python
sns.barplot(x='children',y='charges',data=end)
```

```python
sns.barplot(x='smoker',y='charges',data=end)
```

```python
X=end.drop('charges',axis='columns')
```

```python
X
```

```python
Y=end.charges
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.7 ,random_state=100)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm= LinearRegression()
```

```python
lm.fit(X_train,Y_train)
```

```python
y_pred = lm.predict(X_test)
print(y_pred)
```

```python
df = pd.DataFrame({'Actual': Y_test, 'Predicted': y_pred})
df.plot(kind = 'bar')
plt.ylabel('Profit')
plt.xlabel('Actual and Predicted Datas')
plt.title('Actual and Predicted Data Comparison')
plt.grid()
```

```python
lm.score(X,Y)
```
